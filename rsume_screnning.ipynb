```python
import re
```

```python
def cleanResume(txt):
    cleanTxt = re.sub(r'http\S+\s*', '', txt)
    cleanTxt = re.sub(r'RT|CC', '', cleanTxt)
    cleanTxt = re.sub(r'@\S+',' ',cleanTxt)
    cleanTxt = re.sub(r'#\S+\s',' ', cleanTxt)
    cleanTxt = re.sub(r'[%s]' % re.escape(r"""!"#$%&'()*+,-./:;<=>?@[]^_`{|}~"""), ' ', cleanTxt)
    cleanTxt = re.sub(r'[^\x00-\x7f]', ' ', cleanTxt)
    cleanTxt = re.sub(r'\s+', ' ', cleanTxt)


    return cleanTxt
```

```python
cleanResume("My #### website is like this http://helloworld and acess it @gmain.com")
```

```python
df['Resume'] = df['Resume'].apply(lambda x:cleanResume(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words = 'english')

tfidf.fit(df['Resume'])
requiredText = tfidf.transform(df['Resume'])
```

```python
df
```

```python
requiredText
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(requiredText, df['Category'], test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier 
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

clf = OneVsRestClassifier(KNeighborsClassifier()) #multiclass more than 2oneVsrest
clf.fit(X_train,y_train)
ypred = clf.predict(X_test)
print(ypred)
print(accuracy_score(y_test,ypred))
```

```python
import pickle
pickle.dump(tfidf,open('tfidf.pkl','wb'))  #trained tfidon resume column will now be pickled
pickle.dump(clf, open('clf.pkl', 'wb'))
#thiss pickle filess will be in the main folder 
```
